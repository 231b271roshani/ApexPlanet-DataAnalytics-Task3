```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("Cleaned_Sales_Dataset.csv")
```

```python
# Display Dataset
df.head()
```

```python
# Calculate Funnel Metrics
total_customers = df['Customer_ID'].nunique()
total_orders = df['Order_ID'].nunique()
total_products = df['Quantity'].sum()
total_revenue = df['Total_Sales'].sum()

print("Total Customers:", total_customers)
print("Total Orders:", total_orders)
print("Total Products Sold:", total_products)
print("Total Revenue:", total_revenue)
```

```python
# Create Funnel Data
funnel = pd.DataFrame({
    "Stage": ["Customers", "Orders", "Products Sold", "Revenue"],
    "Value": [total_customers, total_orders, total_products, total_revenue]
})

funnel
```

```python
# Plot Funnel Chart
plt.figure(figsize=(8,5))

plt.barh(funnel["Stage"], funnel["Value"])

plt.title("Sales Funnel Analysis")
plt.xlabel("Value")
plt.ylabel("Stages")

plt.show()
```

```python
# Conversion Rate
conversion_rate = (total_orders / total_customers) * 100

print("Conversion Rate:", round(conversion_rate,2), "%")
```

```python
# Average Revenue Per Customer
revenue_per_customer = total_revenue / total_customers

print("Average Revenue Per Customer:", round(revenue_per_customer,2))
```

```python
# Average Products Per Order
avg_products = total_products / total_orders

print("Average Products Per Order:", round(avg_products,2))
```

```python
# Customer Drop-off
drop_off = total_customers - total_orders

print("Customer Drop-off:", drop_off)
```

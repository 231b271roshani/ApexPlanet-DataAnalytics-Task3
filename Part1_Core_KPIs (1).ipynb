#Define Core KPIs


```python
#Total Revenue
total_revenue = df['Total_Sales'].sum()

print("Total Revenue :", total_revenue)
```

```python
#Total Orders
total_orders = df['Order_ID'].nunique()

print("Total Orders :", total_orders)
```

```python
# Total Customers
total_customers = df['Customer_ID'].nunique()

print("Total Customers :", total_customers)
```

```python
#Average Order Value
average_order_value = total_revenue / total_orders

print("Average Order Value :", round(average_order_value,2))
```

```python
# Total Quantity Sold
total_quantity = df['Quantity'].sum()

print("Total Quantity Sold :", total_quantity)
```

```python
#Average Quantity Per Order
avg_quantity = df['Quantity'].mean()

print("Average Quantity :", round(avg_quantity,2))
```

```python
#Highest Revenue Category
highest_category = df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)

print(highest_category)
```

```python
#Highest Revenue City
highest_city = df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False)

print(highest_city)
```

```python
# Top 5 Products
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)

print(top_products)
```

```python
#Customer Lifetime Value
clv = df.groupby('Customer_ID')['Total_Sales'].sum().mean()

print("Customer Lifetime Value :", round(clv,2))
```

```python
#Repeat Customers
repeat_customers = df.groupby('Customer_ID').size()

repeat_customers = repeat_customers[repeat_customers>1].count()

print("Repeat Customers :", repeat_customers)
```

```python
#Repeat Customer Rate
repeat_rate = (repeat_customers/total_customers)*100

print("Repeat Customer Rate :", round(repeat_rate,2),"%")
```

```python
print("="*50)
print("          CORE KPIs")
print("="*50)

print(f"Total Revenue           : {total_revenue:,.2f}")
print(f"Total Orders            : {total_orders}")
print(f"Total Customers         : {total_customers}")
print(f"Average Order Value     : {average_order_value:.2f}")
print(f"Total Quantity Sold     : {total_quantity}")
print(f"Average Quantity        : {avg_quantity:.2f}")
print(f"Customer Lifetime Value : {clv:.2f}")
print(f"Repeat Customer Rate    : {repeat_rate:.2f}%")
```
